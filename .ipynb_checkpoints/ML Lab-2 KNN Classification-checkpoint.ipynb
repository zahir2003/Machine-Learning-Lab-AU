```python
# One-Hot Encode the categorical column
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['diagnosis'] = encoder.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
# Select all float64 columns
int_feature = df.select_dtypes(include=["float64"]).columns

# Print the total number of float features
print("Total number of float features:", len(int_feature))  # or int_feature.shape[0]

# Print the names of the float features
print("Float feature names:", int_feature)
```

```python
for col in int_feature:
   plt.figure(figsize=(14,7))
   plt.subplot(121)
   sns.boxplot(df[col])
   plt.title(col)
    
   plt.subplot(122)
   sns.histplot(df[col], kde=True,stat="density", kde_kws=dict(cut=3))
   plt.title(col)
plt.show()
```

```python
import numpy as np

# Create a copy of the original DataFrame
new_df_cap = df.copy()

# Loop through each float feature
for col in int_feature:
    # Calculate the 25th and 75th percentiles
    percentile25 = df[col].quantile(0.25)
    percentile75 = df[col].quantile(0.75)

    # Calculate IQR
    iqr = percentile75 - percentile25

    # Calculate upper and lower limits
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr

    # Apply capping for outliers
    new_df_cap[col] = np.where(
        new_df_cap[col] > upper_limit, 
        upper_limit,  # Cap the value to upper_limit if it's higher
        np.where(
            new_df_cap[col] < lower_limit, 
            lower_limit,  # Cap the value to lower_limit if it's lower
            new_df_cap[col]  # Otherwise, leave the value unchanged
        )
    )

print("Outliers capped using IQR method.")
```

```python
for col in int_feature:
   plt.figure(figsize=(14,7))
   plt.subplot(121)
   sns.boxplot(new_df_cap[col])
   plt.title(col)
    
   plt.subplot(122)
   sns.histplot(new_df_cap[col], kde=True,stat="density", kde_kws=dict(cut=3))
   plt.title(col)
plt.show()
```

```python
X = new_df_cap.drop(columns=['diagnosis'])
y = new_df_cap['diagnosis']
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix



# Make predictions
y_pred = neigh.predict(X_test)  # Prediction on test data

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Overall Accuracy: {accuracy:.4f}")
print(confusion_matrix(y_test, y_pred))
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test, y_pred),annot=True)
```
